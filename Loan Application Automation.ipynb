Importing the library and the dataset

```python
import pandas as pd
```

Removing unnecessary columns from the dataset.

```python
data1 = data[['Term', 'NoEmp', 'NewExist','CreateJob', 'RetainedJob','UrbanRural','RevLineCr',
       'LowDoc','DisbursementGross','BalanceGross', 'MIS_Status','GrAppv', 'SBA_Appv']]
data1
```

```python
data1.info()
```

```python
data1.isnull().sum()
```

Filling in Missing values

```python
# Filling Missing Values using mode as all are categorical columns
col = ['NewExist','RevLineCr','LowDoc','MIS_Status']

for i in col:
    data1[i].fillna(data1[i].mode()[0],inplace=True)

data1.isnull().sum()
```

```python
data1.head()
```

```python
# Formatting Amount Columns
col_fmt = ['DisbursementGross','BalanceGross','GrAppv','SBA_Appv']
for i in col_fmt:
    data1[i]= data1[i].str.replace('$', '').str.replace(',', '').astype(float)
    
```

```python
data1.head()
```

```python
#for changing output column to numerical column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

data1['MIS_Status'] = le.fit_transform(data1['MIS_Status'])
data1.head()
```

```python
# Creating dummies for categorical column
cat_col = ['RevLineCr','LowDoc']
final_Data = pd.get_dummies(data1,columns=cat_col,drop_first=True)
final_Data
```

Preprocessing done. Now Heading for Model development

```python
# defining x and y
```

```python
y=final_Data[['MIS_Status']]
x=final_Data.drop(columns=['MIS_Status'])
```

# Logistic Regression Model

```python
# Splitting the data into train and test data

from sklearn.model_selection import train_test_split
x_trn,x_test,y_trn,y_test = train_test_split(x,y,train_size=0.8,random_state=1234)
len(x_trn),len(x_test),len(y_trn),len(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
Lr = LogisticRegression()
Model = Lr.fit(x_trn,y_trn)
Model
```

```python
# Predicting Y with model

y_test['Pred_MIS'] = Model.predict(x_test)

print(y_test)
```

```python
# confusion matrix,classification report and accuracy
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(accuracy_score(y_test['Pred_MIS'],y_test['MIS_Status'])*100,'%')
```

```python
print(classification_report(y_test['Pred_MIS'],y_test['MIS_Status']))
```
